```python
import csv
import json

def jsonl_to_csv(jsonl_filename, csv_filename):
    with open(jsonl_filename, 'r') as jsonl_file:
        with open(csv_filename, 'w', newline='') as csv_file:
            csv_writer = csv.writer(csv_file)
            
            # Write header
            header_written = False
            
            for line in jsonl_file:
                data = json.loads(line[:-2])
                
                # Write header using keys from the first JSON object
                if not header_written:
                    header = data.keys()
                    csv_writer.writerow(header)
                    header_written = True
                
                # Write data rows
                csv_writer.writerow(data.values())
                print(".", end="")

if __name__ == "__main__":
    jsonl_filename = "/Users/denssumesh/Downloads/data.jsonl"  # Replace with your JSONL file's name
    csv_filename = "data.csv"      # Replace with the desired CSV file's name
    
    jsonl_to_csv(jsonl_filename, csv_filename)
    print("Conversion complete.")
```

```python
from numpy import Infinity
import pandas as pd
import json
import ast

# Define the input and output file paths
input_file = '/Users/denssumesh/Documents/GitHub/ai-editor/server-python/data.csv'
output_file = '/Volumes/Backup/Output/output.json'

# Function to process the JSON array and return processed items
def process_json_array(json_array):
    processed_items = []
    for item in json_array:
        prompt = item['f1']
        completion = item['f2']
        if prompt != None and completion != None: 
            processed_item = {
                'prompt': prompt + "\n\n###\n\n",
                'completion': " " + completion + "###"
            }
            if len(prompt) + len(completion) > 11000:
                continue
            processed_items.append(processed_item)
    return processed_items

# Read and process the CSV file in chunks
chunk_size = 1000  # Adjust this based on your memory constraints
processed_data = []
with open(output_file, 'w') as json_file:
    json_file.write("[")

for chunk in pd.read_csv(input_file, chunksize=chunk_size):
    for index, row in chunk.iterrows():
        processed_item = []
        losestContent = ""
        closestCardHTML = ""

        closestCardLengthDiff = Infinity
        
        json_agg = json.loads(json.dumps(ast.literal_eval(row['json_agg'])))
        processed_data.extend(process_json_array(json_agg))
        
        prompt = row['content']
        completion = row['card_html']
        if len(prompt) + len(completion) > 11000:
            continue
        processed_item = {
            'prompt': prompt,
            'completion': completion
        }
        processed_data.append(processed_item)
        print(index)
        for item in processed_data:
            cardLength = len(item['completion'])
            lengthDiff = abs(cardLength - 4300)
            if cardLength <= 4300 and lengthDiff < closestCardLengthDiff:
                closestCardHTML = item['completion']
                closestContent = item['prompt']
                closestCardLengthDiff = lengthDiff
                

        with open(output_file, 'a') as json_file:
            string  = json.dumps({
                'prompt': closestContent,
                'completion': closestCardHTML
            })
            json_file.write(string.replace("[", "").replace("]", "") + ",")
with open(output_file, 'a') as json_file:
    json_file.write("]")

# Write the processed data to the output JSON file


print("Data processing and conversion complete.")
```
